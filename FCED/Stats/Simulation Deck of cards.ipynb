```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_hands():
    deck = [i for i in range(1,49)]
    np.random.choice(deck,replace = False)
    female=[]
    male = []
    for i in range(24):
        rn = np.random.choice(deck,replace = False)
        female.append(rn)
        deck.remove(rn)
    for i in range(24):
        rn = np.random.choice(deck,replace = False)
        male.append(rn)
        deck.remove(rn)
    if len(deck) > 0:
        raise Exception("Deck still have cards")
    return female, male
    
```

```python
def filter_prom(num):
    if num > 13:
        return True 
def simulation():
    f,m = get_hands()
    filtered_m = list(filter(filter_prom,m))
    filtered_f = list(filter(filter_prom,f))
    perc_m = len(filtered_m)/len(m)
    perc_f= len(filtered_f)/len(f)
    diff = perc_m - perc_f
    return diff
```

```python

record = []
for i in range(100):
    record.append(simulation())
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
def dotplot(input_x,width,height ,**args):
 
    # Count how many times does each value occur
    unique_values, counts = np.unique(input_x, return_counts=True)
    
    # Convert 1D input into 2D array
    scatter_x = [] # x values 
    scatter_y = [] # corresponding y values
    for idx, value in enumerate(unique_values):
        for counter in range(1, counts[idx]+1):
            scatter_x.append(value)
            scatter_y.append(counter)
 
    # draw dot plot using scatter()
    plt.figure(figsize=(width,height))
    plt.scatter(scatter_x, scatter_y, **args)
    
```

```python
dotplot(record,20,10)
```
